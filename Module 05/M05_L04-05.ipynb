## Problem 2

(a) Using the data file in chicago_hourly_aug_2015.csv, write out csv files for each hour for August 2015.

```python
import pandas as pd
import numpy as np
df = pd.read_csv('chicago_hourly_aug_2015.csv', skiprows = 6, na_values='    ')
df['sethour'] = df['Time']/100
df['Time'] = np.floor_divide(df['Time'],100)
df['dates'] = pd.to_datetime(df['Date'],format='%Y%m%d') + pd.to_timedelta(df['Time'], unit = 'h')
df.index = df['dates']
df['sethour']=df['sethour'].astype(int)
df['just_dates'] = df['dates'].dt.date
for i in range(24):
    data = df[df['sethour'] == i]
    name = 'hour_data/'+'Hour_' + str(i) + '.csv'
    a = list(df) 
    data.to_csv(name, sep = ",", columns = a, na_rep = 'NaN')
   
```
