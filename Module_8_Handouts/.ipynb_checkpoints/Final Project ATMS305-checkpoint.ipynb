<h2> Map of this month Earthquake </h2>
<h3>Importing earthquake data obtained from USGS for the last 30-days ago. The data is ever-changing as it takes live-feed data from USGS  </h3>

```python
import warnings
warnings.filterwarnings('ignore')
import pandas as pd
import numpy as np
import matplotlib
import matplotlib.pyplot as plt
matplotlib.style.use('ggplot')
import datetime as dt
import os
from mpl_toolkits.basemap import Basemap
%matplotlib inline
```

```python
import pandas as pd
eq = pd.read_csv("http://earthquake.usgs.gov/earthquakes/feed/v1.0/summary/all_month.csv")
eq["DateTime"] = (pd.to_datetime(eq["time"]))
eq["Date"] = (pd.to_datetime(eq["time"])).dt.date
eq["Time"] = (pd.to_datetime(eq["time"])).dt.time
eq["Date"] = eq["Date"].astype("datetime64")
listdf = eq['mag'].count()
eq
```

```python
print("There is a %d numbers of earthquake in the last 30 days" %listdf)
```

```python
m = Basemap(projection="mill")
```

```python
x,y = m([longs for longs in eq["longitude"]],
         [lats for lats in eq["latitude"]])
fig = plt.figure(figsize=(20,20))
plt.title("Earthquake in the last 30-days")
m.scatter(x,y, s = 100, c = eq["mag"] )
m.drawcoastlines()
m.drawmapboundary()
m.drawcountries()
m.bluemarble()
m.fillcontinents(color='lightsteelblue',lake_color='skyblue')
plt.colorbar()
plt.show()
```

<h2>Frequency of earthquake in terms of size. </h2>
<p>We want to find the number of earthquake in terms of size to see what can we plot from the data obtained </p>

```python
maxeq = eq['mag'].max()
mineq = eq['mag'].min()
counteq = eq["mag"].count()
eq["mag"].describe()
```

```python
print("Out of the total earthquake of %d, the maximum magnitude is %.2f Richter scale, while the minimum is %.2f Richter scale" %(counteq, maxeq,mineq))
```

```python
eq1 = eq[eq["mag"] <= 1]
eq2 = eq[(eq["mag"] <= 2) & (eq["mag"] > 1)]
eq3 = eq[(eq["mag"] <= 3) & (eq["mag"] > 2)]
eq4 = eq[(eq["mag"] <= 4) & (eq["mag"] > 3)]
eq5 = eq[(eq["mag"] <= 5) & (eq["mag"] > 4)]
eq6 = eq[(eq["mag"] <= 6) & (eq["mag"] > 5)]
eq7 = eq[(eq["mag"] <= 7) & (eq["mag"] > 6)]
eq8 = eq[(eq["mag"] <= 8) & (eq["mag"] > 7)]
```

```python
frame = {'Frequency': [len(eq1), len(eq2),len(eq3),len(eq4),len(eq5),len(eq6),len(eq7),len(eq8)],'Magnitude': [1, 2,3,4,5,6,7,8]}
Freqf = pd.DataFrame(data=frame)
Freqf['Percentage'] = round((Freqf['Frequency']/Freqf['Frequency'].sum())*100,2)
Freqf
```

```python
#Freqf["Frequency"].plot.bar()
plt.bar(Freqf["Magnitude"],Freqf["Frequency"], color = "Maroon")
plt.xlabel('Magnitude')
plt.ylabel('Frequency')
plt.title('Earthquake Magnitude vs Frequency')
plt.grid(color='black', linestyle='--', linewidth=0.5, alpha = 0.2)
plt.show()
```

Increase in magnitude reflect in decrease in frequency. Earthquake of a size 6 is 10 times more likely than 7. The oddity in this data is in magnitude 4 earthquake, where it is significantly less than 3 and 5. This anomaly is due to insufficient data as we only take data from the last 30 days. 

<h2> Correlation of Depth vs Magnitude <h2>

```python
shallow = len(eq[eq.depth < 70])
intermediate = len(eq[(eq.depth > 70) & (eq.depth < 300)])
deep = len(eq[eq.depth > 300])
percshallow = (shallow/(shallow + intermediate + deep))*100
percinter = (intermediate/(shallow + intermediate + deep))*100
percdeep = (deep/(shallow + intermediate + deep))*100
#print "He's %d pounds heavy." % my_weight
print("%.2f percent of earthquakes are shallow." %percshallow)
print("%.2f percent of earthquakes are intermediate." %percinter)
print("%.2f percent of earthquakes are deep." %percdeep)
```

We can observe that most of the earthquake that occur in the last 30 days are shallow, followed by intermediate, then deep - which fit the geological theory on earthquake.

```python
plt.scatter(eq["mag"],eq["depth"])
plt.xlabel("Magnitude")
plt.ylabel("Depth (in meters)")
plt.title("Magnitude vs Depth")
plt.show()
```

From the plot above, there is something interesting occur in magnitude 4 - 5. We can see a majority of deep and intermediate earthquake occur in this region. We will plot the map to see what can we deduce from the data above. We will also add plate boundaries to observe the location of these earthquakes

```python
df2 = eq[(eq.mag >= 4) & (eq.mag <= 5)]
m = Basemap(projection="mill")
x,y = m([longs for longs in df2["longitude"]],
         [lats for lats in df2["latitude"]])
fig = plt.figure(figsize=(20,20))
plt.title("Intermediate and Deep Earthquake of Mag. 4 - 5")
m.scatter(x,y, s = 100, c = df2["mag"] )
m.drawcoastlines()
m.drawmapboundary()
m.readshapefile('tectonicplates-master/PB2002_plates', 
                name='tectonic_plates', 
                drawbounds=True, 
                color='red')

#download the readshapefile from https://github.com/fraxen/tectonicplates for tectonic plate data. 
m.drawcountries()
m.bluemarble()
m.fillcontinents(color='lightsteelblue',lake_color='skyblue')
plt.colorbar()
plt.show()
```

The data shows that the earthquake of this size focus around subduction-zone region. It is not surprising as major earthquake usually occur around subduction zone. 

What is interesting is that magnitude 4 - 5 earthquake are the range for deep and intermediate depth earthquake. This is due to the weakening of the tectonic slab due to it being subducted under another tectonic plate. As the slab gets subducted deeper, the heat from the mantle would melt the slab hence reducing the strength of the slab. That is why earthquake in this region rarely exceed the magnitude 5 range. 

<h3> Correlation of Date vs Frequency </h3>

```python
eq['mag'].groupby(eq["Date"].dt.day).count().plot(kind="bar", color = 'maroon')
plt.xlabel('Day')
plt.ylabel('Frequency')
plt.title('Earthquake Date vs Frequency')
plt.grid(color='black', linestyle='--', linewidth=0.5, alpha = 0.2)
plt.show()
```

Earthquake are evenly distributed throughout the whole month. 

<h3> Ring of Fire </h3>

```python
rof_lat = [-61.270, 56.632]
rof_long = [-70, 120]
ringoffire = eq[((eq.latitude < rof_lat[1]) & (eq.latitude > rof_lat[0]) & ~((eq.longitude < rof_long[1]) & (eq.longitude > rof_long[0])))]
percentage_of_total_eq = (ringoffire['mag'].count() / eq.mag.count())*100
print("The percentage of total number of earthquake that lies in the ring of fire is %.2f %%"%percentage_of_total_eq )
#ringoffire
```

```python
m = Basemap(projection="mill" )
x,y = m([longs for longs in ringoffire["longitude"]],
         [lats for lats in ringoffire["latitude"]])
fig2 = plt.figure(figsize=(20,20))
plt.title("Earthquakes in the Ring of Fire Area")
m.scatter(x,y, s = 15, c = "maroon")
m.drawcoastlines()
m.readshapefile('tectonicplates-master/PB2002_plates', 
                name='tectonic_plates', 
                drawbounds=True, 
                color='red')
m.drawmapboundary()
m.drawcountries()
m.fillcontinents(color='lightsteelblue',lake_color='skyblue')
plt.show()
```

```python
maxeq = ringoffire['mag'].max()
mineq = ringoffire['mag'].min()
counteq = ringoffire["mag"].count()
print("Out of the total earthquake of %d, the maximum magnitude is %.2f Richter scale, while the minimum is %.2f Richter scale" %(counteq, maxeq,mineq))
```
